```python
import numpy as np
import matplotlib.pyplot as plt

# ----- dados sintticos -----
np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

# matriz com coluna de 1s (intercepto)
x_b = np.c_[np.ones((len(x), 1)), x]

# ----- hiperparmetros -----
eta = 0.1
n_iterations = 10
m = len(x)

# ----- inicializao -----
theta = np.random.randn(2, 1)

# ----- figura/axes -----
fig, ax = plt.subplots(figsize=(12, 6))

# espalhar pontos originais
ax.scatter(x, y, alpha=0.6, label="Dados")

# para a linha ficar lisa: ordenar x e montar x_b ordenado
order = np.argsort(x.ravel())
x_sorted = x[order]
x_b_sorted = np.c_[np.ones((len(x_sorted), 1)), x_sorted]

# ----- treino por gradiente descendente com plots -----
for iteration in range(n_iterations):
    # gradiente e atualizao
    gradients = (2/m) * x_b.T.dot(x_b.dot(theta) - y)
    theta = theta - eta * gradients

    # previso usando theta atual (na ordem ordenada)
    y_hat_sorted = x_b_sorted.dot(theta)

    # alpha de 0.2 a 1.0 (sem estourar)
    frac = iteration / (n_iterations - 1) if n_iterations > 1 else 1.0
    alpha = 0.2 + 0.8 * frac  # sempre entre 0.2 e 1.0

    # legenda s na ltima iterao
    lbl = "Modelo final" if iteration == n_iterations - 1 else "_nolegend_"
    ax.plot(x_sorted, y_hat_sorted, alpha=alpha, label=lbl)

# ajustes finais
ax.set_title("Evoluo da Regresso por Gradiente Descendente")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend(loc="upper left")

plt.tight_layout()
plt.show()
```

```python
# codig mais completo chat com legenda 
# parmetros
eta = 0.1  # taxa de aprendizado
n_iterations = 10
m = len(x)

# inicializa theta aleatrio
theta = np.random.randn(2, 1)

# criar figura
fig, ax = plt.subplots(figsize=(25, 8))

# plotar dados originais
ax.scatter(x, y, color='blue', alpha=0.6, label='Dados Originais')

# gradiente descendente
for iteration in range(n_iterations):
    # clculo do gradiente
    gradients = 2/m * x_b.T.dot(x_b.dot(theta) - y)
    theta = theta - eta * gradients  # atualizao do theta
    
    # reta prevista (usando theta atual)
    y_predict = x_b.dot(theta)
    
    # desenhar a reta da iterao atual (com transparncia)
    ax.plot(x, y_predict, linewidth=2, alpha=0.3,
            label=f'Iterao {iteration + 1}')

# destacar a ltima reta (modelo final)
ax.plot(x, y_predict, color='red', linewidth=3, label='Modelo Final')

# legendas e ttulo
ax.set_title('Evoluo da Regresso Linear por Gradiente Descendente', fontsize=18)
ax.set_xlabel('Varivel X')
ax.set_ylabel('Varivel Y')
ax.legend(loc='upper left', fontsize=10, ncol=2)

plt.show()
```

### Ridge Regression

#### A Ridge Regression  uma verso modificada da Regresso Linear, criada para evitar overfitting.
#### Ela adiciona uma penalizao (chamada de termo de regularizao) 
#### sobre os coeficientes do modelo  ou seja, ela tenta manter os pesos pequenos para o modelo no 
#### ficar maluco se os dados tiverem rudo.


```python
# --------------------------------------------
#  Exemplo prtico de Ridge Regression (Regresso de Crista)
# --------------------------------------------

# Importa as bibliotecas necessrias
from sklearn.linear_model import Ridge                   # Modelo Ridge (regresso com regularizao L2)
from sklearn.preprocessing import PolynomialFeatures      # Cria atributos polinomiais
from sklearn.pipeline import make_pipeline                # Facilita encadear etapas (pipeline)
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------
#  1 Gerar dados de exemplo (com rudo)
# --------------------------------------------

np.random.seed(42)  # garante reprodutibilidade
x = 6 * np.random.rand(100, 1) - 3   # valores de X entre -3 e 3
y = 0.5 * x**2 + x + 2 + np.random.randn(100, 1)  # relao quadrtica com rudo aleatrio

# --------------------------------------------
#  2 Criar um modelo com pipeline
# --------------------------------------------

# A pipeline encadeia duas etapas:
# - PolynomialFeatures(degree=15): adiciona colunas x, x, ..., x (modelo altamente complexo)
# - Ridge(alpha=1): aplica regresso linear com regularizao L2 (controla overfitting)

ridge_model = make_pipeline(PolynomialFeatures(degree=15), Ridge(alpha=1))

# Treina o modelo (ajusta os coeficientes )
ridge_model.fit(x, y)

# --------------------------------------------
#  3 Fazer previses com novos valores de X
# --------------------------------------------

# Gera 200 valores igualmente espaados entre -3 e 3
x_new = np.linspace(-3, 3, 200).reshape(-1, 1)

# Usa o modelo treinado para prever os valores correspondentes de Y
y_pred = ridge_model.predict(x_new)

# --------------------------------------------
#  4 Visualizar o resultado
# --------------------------------------------

plt.figure(figsize=(10, 6))

# Pontos reais dos dados (azuis)
plt.scatter(x, y, color="blue", alpha=0.5, label="Dados reais")

# Curva predita pelo modelo Ridge (vermelha)
plt.plot(x_new, y_pred, color="red", linewidth=2, label="Ridge Regression (=1)")

# Legendas e ttulo
plt.legend()
plt.title("Ridge Regression - Regularizao L2", fontsize=14)
plt.xlabel("Varivel X")
plt.ylabel("Varivel Y")

plt.show()
```

```python
# 3 impactos diferentes geram no conteudo

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

# --- gerar dados de exemplo (caso no tenha) ---
np.random.seed(42)
x = 2 * np.random.rand(100, 1) - 1
y = 0.5 * x**2 + x + np.random.randn(100, 1)

# --- criar figura ---
fig, ax = plt.subplots(figsize=(25, 8))

#  plota os pontos reais uma vez
plt.scatter(x, y, color='blue', alpha=0.6, label="Dados")

# --- testar 3 valores diferentes de alpha (regularizao Ridge) ---
for alpha in [0, 1e-5, 1]:
    # cria polinmio de grau 10
    poly_features = PolynomialFeatures(degree=10, include_bias=False)
    x_poly = poly_features.fit_transform(x)

    # cria sequncia suave de X (para desenhar curva)
    x_seq = np.linspace(x.min(), x.max(), 300).reshape(-1, 1)
    x_seq_transf = poly_features.fit_transform(x_seq)

    # cria e treina o modelo Ridge
    ridge_reg = Ridge(alpha=alpha, solver="cholesky")
    ridge_reg.fit(x_poly, y)

    # plota a curva resultante
    plt.plot(
        x_seq,
        ridge_reg.predict(x_seq_transf),
        linewidth=2,
        label=f"Ridge ={alpha}"
    )

# --- acabamento ---
ax.legend(title="Fora da regularizao ")
ax.set_title("Impacto da regularizao Ridge () em modelos polinomiais", fontsize=16)
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```
